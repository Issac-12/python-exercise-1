```python
from bs4 import BeautifulSoup

import pandas as pd

import requests

#soup operation 

response = requests.get("https://ravindrababuravula.com/courses.php")

soup = BeautifulSoup(response.text, "html.parser")

print(soup)

#make temp df

df = pd.DataFrame()
df2 = pd.DataFrame()

#loop courses in df
for i in range(0,3):
    table = soup.find_all("a",attrs={'class':"title"})[i]

    
    text = [(''.join(table.stripped_strings))]
    
    df = df.append(text)
    
    i=i+1
    

#loop descriptopion in df2
for j in range (0,3):
    table2 = soup.find_all("div",attrs={'class':"description"})[j]

    text2 = [''.join(table2.stripped_strings)]
    
    df2 = df2.append(text2)
    
    j=j+1

#combine df and df2 to df3, change column name 

df3 = pd.concat([df, df2], axis=1) 
df3.columns =['class', 'Description']
df3
```
